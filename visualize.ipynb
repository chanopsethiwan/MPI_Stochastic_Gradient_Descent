```python
import pickle

file_path = 'results.pkl'
with open(file_path, 'rb') as file:
    results = pickle.load(file)

# # Now 'data' contains the deserialized Python object
# print(re)
```

```python
result = results[0]
result.keys()
```

```python
import matplotlib.pyplot as plt

for result in results:
    train_err = result['train_err']
    test_err = result['test_err']
    act_func = result['activation_func']
    batch_size = result['batch_size']

    plt.plot(range(len(train_err)), train_err, label="Train RMSE")
    plt.plot(range(len(test_err)), test_err, label="Test RMSE")
    plt.xlabel("Epoch")
    plt.ylabel("RMSE")
    plt.legend()
    plt.title("Training vs Testing Error")
    plt.show()
```
